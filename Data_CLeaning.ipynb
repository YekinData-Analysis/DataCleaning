SO many Unnecessary Str attached to our value. 

```python
#First of All We need t

import re

#Creating def fucntion that will take off the Unnecessary string
def preprocessor(text):
             text=re.sub('<[^>]*>','',text)
             emojis=re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)',text)
             text=re.sub('[\W]+',' ',text.lower()) +\
                ' '.join(emojis).replace('-','')
             return text   
```

```python
#Now time to Apply to Each Columns

df['Track Name']=df['Track Name'].apply(preprocessor)
```

```python
df['Artist Name']=df['Artist Name'].apply(preprocessor)
```

```python
df['Genres']=df['Genres'].apply(preprocessor)
```
